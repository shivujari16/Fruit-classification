```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\fruit_data.csv")
df
```

```python
df.isnull().sum()
```

```python
df.fruit_subtype.unique()
```

```python
df.fruit_subtype = pd.Categorical(df.fruit_subtype,['granny_smith', 'mandarin', 'braeburn', 'golden_delicious',
       'cripps_pink', 'spanish_jumbo', 'selected_seconds', 'turkey_navel',
       'spanish_belsan', 'unknown'],ordered=True)
df.fruit_subtype = df.fruit_subtype.cat.codes
```

```python
df
```

```python
X = df.drop("fruit_name",axis = 1)
y = df["fruit_name"]
```

# Feature selection

```python
from  sklearn.feature_selection import SelectKBest
from  sklearn.feature_selection import chi2

feature = SelectKBest(score_func = chi2 , k = 5)
fit = feature.fit(X,y)
score = pd.DataFrame(fit.scores_)
column = pd.DataFrame(X.columns)

ft = pd.concat([column,score],axis = 1)
ft
```

```python
plt.figure(figsize = (15,15))
cor = df.corr()
top = cor.index
sns.heatmap(df[top].corr(),annot = True,cmap = 'RdYlGn')
```

```python
X = df[["width","height","color_score"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 51)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# feature scalling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_sc = sc.transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc = pd.DataFrame(X_train_sc, columns=X.columns)
X_test_sc = pd.DataFrame(X_test_sc, columns=X.columns)
```

# SVC

```python
from sklearn.svm import SVC
svc_linear = SVC(kernel='linear')
svc_linear.fit(X_train,y_train)
svc_linear.score(X_test,y_test)
```
